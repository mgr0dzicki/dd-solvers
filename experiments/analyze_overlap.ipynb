```python
import pandas as pd
```

```python
df1 = pd.read_csv("overlapping_asm_results.csv")
df2 = pd.read_csv("overlapping_asm_results_9.csv")
df = pd.concat([df1, df2])
df
```

```python
def get_order(row):
    h = 1 / row["n"]
    H = 1 / row["N"]
    HH = 1 / row["NN"]
    delta = row["overlap_n"] / row["n"]
    return HH * HH / H / max(h, delta)

df["estimate"] = df.apply(get_order, axis=1)
df["ratio"] = df["condition number"] / df["estimate"]
df[(df.N == df.NN) & (df.n == 512)].pivot_table(values="condition number", index="N", columns="overlap_n")
```
