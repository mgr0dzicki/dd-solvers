```python
import pandas as pd
```

```python
df = pd.read_csv("overlapping_schwarz_results.csv")
df
```

```python
def get_order(row):
    h = 1 / row["n"]
    H = 1 / row["N"]
    HH = 1 / row["NN"]
    delta = row["overlap_n"] / row["n"]
    return HH * HH / H / max(h, delta)

df["estimate"] = df.apply(get_order, axis=1)
df["ratio"] = df["condition number"] / df["estimate"]
```

```python
df["solver"] = df["solver"] + "+" + df["preconditioner"]
```

```python
df[(df.NN == 16) & (df.n == 256) & (df.coefficient == "constant")].pivot_table(values="iterations", index=["N", "overlap_n"], columns="solver", aggfunc="min")
```

```python
df[(df.NN == 16) & (df.n == 256) & (df.coefficient == "coarse checkerboard")].pivot_table(values="iterations", index=["N", "overlap_n"], columns="solver", aggfunc="min")
```

```python
df[(df.NN == 16) & (df.n == 256) & (df.coefficient == "solvers checkerboard")].pivot_table(values="iterations", index=["N", "overlap_n"], columns="solver", aggfunc="min")
```
