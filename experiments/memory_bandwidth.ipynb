```python
import torch
from benchmark_utils import timeit
```

```python
a = torch.empty((2_000_000_000), dtype=torch.float64, device='cuda')
```

```python
t = timeit(lambda: a.clone(), warmup_iters=0, iters=1, repetitions=100)
t
```

```python
total_bytes = a.numel() * a.element_size() * 2  # read + write
bw = total_bytes / (t / 1000) / 1e9
print(f'Bandwidth: {bw:.2f} GB/s')
```
