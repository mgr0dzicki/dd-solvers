```python
from dd_solvers import *
from experiments import *
import matplotlib.pyplot as plt
```

```python
OFFSET = 0.03
COARSE_STYLE = {
    "color": "black",
    "linewidth": 3,
    "capstyle": "round",
}
SOLVERS_STYLE = {
    "color": "black",
    "linewidth": 1,
}
FINE_STYLE = {
    "color": "#cccccc",
    "linewidth": 1,
}
```

```python
mesh_family_triangle = MeshFamily.refinements(
    base=Mesh2D.unit_square_uniform(1, 1),
    n=4,
)
mesh_family_square = MeshFamily.uniform_quadrilateral(
    d=2,
    n=4,
)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
for ax in axes:
    ax.set_xlim(-OFFSET, 1 + OFFSET)
    ax.set_ylim(-OFFSET, 1 + OFFSET)
    ax.set_axis_off()

mesh_family_triangle[3].plot(axes[0], **FINE_STYLE)
mesh_family_triangle[2].plot(axes[0], **SOLVERS_STYLE)
mesh_family_triangle[1].plot(axes[0], **COARSE_STYLE)

mesh_family_square[4].plot(axes[1], **FINE_STYLE)
mesh_family_square[3].plot(axes[1], **SOLVERS_STYLE)
mesh_family_square[1].plot(axes[1], **COARSE_STYLE)

fig.savefig("../docs/thesis-media/mesh_examples.svg", bbox_inches="tight")
```
